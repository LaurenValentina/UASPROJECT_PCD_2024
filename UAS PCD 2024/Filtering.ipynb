```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def mean_filter(image):
    copy_image = image.copy().astype(float)
    m, n = copy_image.shape
    output = np.empty_like(copy_image)
    
    for row in range(1, m-1):
        for col in range(1, n-1):
            total = (
                copy_image[row-1, col-1] + copy_image[row-1, col] + copy_image[row-1, col+1] +
                copy_image[row, col-1] + copy_image[row, col] + copy_image[row, col+1] +
                copy_image[row+1, col-1] + copy_image[row+1, col] + copy_image[row+1, col+1]
            )
            output[row, col] = total / 9
    
    # Mengonversi kembali ke tipe data yang sesuai (uint8)
    output = output.astype(np.uint8)
    return output
```

```python
# Membaca gambar dalam mode grayscale
img = cv2.imread('lauren.jpg', cv2.IMREAD_GRAYSCALE)

# Menerapkan mean filter
mean_filtered = mean_filter(img)

# Menampilkan hasil
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original Image')
plt.subplot(1, 2, 2), plt.imshow(mean_filtered, cmap='gray'), plt.title('Mean Filtered Image')
plt.show()
```
