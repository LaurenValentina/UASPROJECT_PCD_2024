# deteksi tepi

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Membaca gambar
img = cv2.imread('lauren.jpg')
```

```python
img_canny = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Deteksi tepi menggunakan Canny
canny = cv2.Canny(img, 100, 100)

# TAMPILAN MENGGUNAKAN FIGUR DAN XIS (FUNGSI MATPLOTLIB)
fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

# Tampilkan gambar asli
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title ('Original Image ')

# Tampilkan hasil deteksi daun
ax[1].imshow(canny, cmap='gray')
ax[1].set_title('Canny Edge Detection')
```

```python
# Deteksi kontur
contours, _ = cv2.findContours(canny, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Membuat salinan gambar asli untuk menggambar kontur
contour_img = cv2.cvtColor(img_canny, cv2.COLOR_GRAY2BGR)
cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)

# TAMPILAN MENGGUNAKAN FIGUR DAN XIS (FUNGSI MATPLOTLIB)
fig, axs = plt.subplots(1,3, figsize = (10,10))
ax = axs.ravel()

# Tampilkan gambar asli
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title ('Original Image ')

# Tampilkan hasil deteksi daun
ax[1].imshow(canny, cmap='gray')
ax[1].set_title('Canny Edge Detection')

# Tampilkan hasil deteksi daun
ax[2].imshow(contour_img, cmap='gray')
ax[2].set_title('Contours Detection')
```
