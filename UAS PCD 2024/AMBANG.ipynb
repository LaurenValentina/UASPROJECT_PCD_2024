```python
import cv2 #menjalankan fungsi" library open cv 
import numpy as np #menjalankan fungsi array untuk mengeluarkan 
import matplotlib.pyplot as plt #menampilkan fungsi visualisasi
```

```python
daun = cv2.imread("daun.jpeg",0) #Membuat variabel daun untuk mengimpor atau membaca file “daun.jpg”
tinggi, lebar= daun.shape  #di dalam shape terdapat panjang , tinggi dan dimensi
```

```python
def titeratif(image): #kita menggunakan kata image hanya untuk pendelarasian umum untuk rumua
    tinggi, lebar = image.shape
    t0 = 127
    while True:
        rata_kiri = 0
        rata_kanan = 0
        jum_kiri = 0
        jum_kanan = 0
        for x in range(tinggi):
            for y in range(lebar):
                if image[x, y] <= 127:
                    rata_kiri = rata_kiri + image[x, y]
                    jum_kiri = jum_kiri + 1
                else:
                    rata_kanan = rata_kanan + image[x, y]
                    jum_kanan = jum_kanan + 1

        rata_kiri = rata_kiri / jum_kiri
        rata_kanan = rata_kanan / jum_kanan
        t1 = (rata_kiri + rata_kanan) / 2
        if (t0 - t1) < 1:
            break
        t0 = t1
    return round(t1)
```

```python
titeratif(daun)
```

```python
nilai_ambang = 164 #Menentukan nilai ambang, sebagai acuan menentukan aras
daun_hasil1 = daun.copy() #daun_hasil sebagai variabel yang menampung salinan variabel daun
for x in range (tinggi): #looping pertama untuk 
    for y in range (lebar):
        if daun[x,y] < nilai_ambang :
            daun_hasil1[x,y] = 0
        else:
            daun_hasil1[x,y] = 255
```

```python
# TAMPILAN MENGGUNAKAN FIGUR DAN XIS (FUNGSI MATPLOTLIB)
fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

# Tampilkan gambar asli
ax[0].imshow(daun, cmap='gray')
ax[0].set_title('Citra Asli')

# Tampilkan hasil deteksi daun
ax[1].imshow(daun_hasil1, cmap='gray')
ax[1].set_title('Setelah deteksi daun')
```

```python
def arasjamak(daun , t1, t2) :
    res = daun
    m,n = daun.shape
    for x in range(m):
        for y in range(n):
            if daun[x,y] < nilai_ambang :
                res[x,y] = 0
            else:
                res[x,y] = 225
    return res
```

```python
bentukarasjamak = arasjamak(daun, 185, 200)
```

```python
# TAMPILAN MENGGUNAKAN FIGUR DAN XIS (FUNGSI MATPLOTLIB)
fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

# Tampilkan gambar asli
ax[0].imshow(daun, cmap='gray')
ax[0].set_title('Citra Asli')

# Tampilkan hasil deteksi daun
ax[1].imshow(bentukarasjamak, cmap='gray')
ax[1].set_title('Setelah deteksi daun')
```

```python
daun = cv2.imread("daun.jpeg")
daun_hls = cv2.cvtColor(daun, cv2.COLOR_BGR2HLS)
tinggi, lebar = daun.shape[:2]
```

```python
 for x in range(tinggi):
        for y in range(lebar):
            h,l,s = daun_hls[x,y]
        
        if h < 11 :
            h = 0
        elif h < 32 :
            h = 21
        elif h < 54 :
            h = 43;
        elif h < 116 :
            h = 85;
        elif h < 141 :
            h = 128;
        elif h < 185 :
            h = 170;
        elif h < 202 :
            h = 191;
        elif h < 223 :
            h = 213;
        elif h < 244 :
            h = 234;
        else :
            h = 0;
            
        daun_hls[x,y][0] = h
        
        if s >= 200:
            daun_hls[x,y][2] = 255
        elif s <= 20:
            daun_hls[x,y][2] = 0
        else :
            daun_hls[x,y][2] = 128
            
        if s >= 200:
            daun_hls[x,y][1] = 255;
        elif s <= 20:
            daun_hls[x,y][1] = 0;
        else :
            daun_hls[x,y][1] = 128;
            
daun_res = cv2.cvtColor(daun_hls, cv2.COLOR_HLS2RGB)
```

```python
# TAMPILAN MENGGUNAKAN FIGUR DAN XIS (FUNGSI MATPLOTLIB)
fig, axs = plt.subplots(1,2, figsize = (10,10))
ax = axs.ravel()

# Tampilkan gambar asli
ax[0].imshow(daun, cmap='gray')
ax[0].set_title('Citra Asli')

# Tampilkan hasil deteksi daun
ax[1].imshow(daun_res, cmap='gray')
ax[1].set_title('Setelah deteksi daun')
```
