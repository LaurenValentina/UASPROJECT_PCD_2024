```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Baca gambar
image = cv2.imread('daun.jpeg')

# Konversi gambar dari BGR ke HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

# Definisikan rentang warna untuk mangga (disesuaikan dengan warna mangga)
mango_lower = np.array([10, 100, 100])  # Contoh nilai untuk warna mangga
mango_upper = np.array([30, 255, 255])

# Definisikan rentang warna untuk daun (disesuaikan dengan warna daun)
leaf_lower = np.array([40, 40, 40])  # Contoh nilai untuk warna daun
leaf_upper = np.array([80, 255, 255])

# Buat mask untuk mangga
mango_mask = cv2.inRange(hsv_image, mango_lower, mango_upper)

# Buat mask untuk daun
leaf_mask = cv2.inRange(hsv_image, leaf_lower, leaf_upper)

# Terapkan mask pada gambar asli untuk memisahkan mangga dan daun
mango_segmented = cv2.bitwise_and(image, image, mask=mango_mask)
leaf_segmented = cv2.bitwise_and(image, image, mask=leaf_mask)

# Tampilkan hasil segmentasi
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.title('Gambar Asli')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.subplot(2, 2, 2)
plt.title('Mask Mangga')
plt.imshow(mango_mask, cmap='gray')

plt.subplot(2, 2, 3)
plt.title('Segmentasi Mangga')
plt.imshow(cv2.cvtColor(mango_segmented, cv2.COLOR_BGR2RGB))

plt.subplot(2, 2, 4)
plt.title('Segmentasi Daun')
plt.imshow(cv2.cvtColor(leaf_segmented, cv2.COLOR_BGR2RGB))

plt.show()
```
